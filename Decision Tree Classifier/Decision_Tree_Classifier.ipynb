```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics
```

```python
col_names = ['Id','Sepal Length (cm)', 'Sepal Width (cm)', 'Petal Length (cm)', 'Petal Width (cm)','Species']
# load dataset
iris = pd.read_csv("Iris.csv", header=None, names=col_names) #read csv file
```

```python
iris.head()
```

```python
iris=iris.drop(0)
```

```python
iris.head()
```

```python
feature_cols = ['Sepal Length (cm)', 'Sepal Width (cm)', 'Petal Length (cm)', 'Petal Width (cm)'] #Assign fearure variables
X = iris[feature_cols] # Features
y = iris.Species #Target variable
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test
```

```python
clf = DecisionTreeClassifier(criterion="entropy") #Split tree based on entropy values

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)
```

```python
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

# Visualize the graph
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data, feature_names=feature_cols,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

```python
#Predicting the tagret variable for test values
y_pred = clf.predict(X_test)

#Accuracy of the model- how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
